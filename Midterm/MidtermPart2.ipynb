```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
T1 = 15
n1 = np.arange(0, T1)
x1 = 0.5**n1*np.heaviside(n1, 1)
x1_p = np.tile(x1, 5)

plt.stem(x1_p)
plt.xlabel('n')
plt.ylabel('x1[n]')
plt.title('Periodic signal x1[n]')
plt.show()
```

```python
T2 = 6
n2 = np.arange(0, T2)
d = lambda n: np.heaviside(n, 1) - np.heaviside(n-1, 1)
x2 = d(n2) + 2*d(n2-1) + 4*d(n2-3) + 5*d(n2-4) - 10*d(n2-5)
x2_p = np.tile(x2, 5)

plt.stem(x2_p )
plt.xlabel('n')
plt.ylabel('x2[n]')
plt.title('Periodic signal x1[n]')
plt.show()
```

```python
a1 = np.zeros(T1, dtype = complex)
for k in range(T1):
    a1[k] = np.mean(x1*np.exp(-1j*2*np.pi*k*n1/T1))
print("Coefficients for x1[n]:")
print(a1)

a2 = np.zeros(T2, dtype = complex)
for k in range(T2):
    a2[k] = np.mean(x2*np.exp(-1j*2*np.pi*k*n2/T2))
print("\nCoefficients for x2[n]:")
print(a2)
```

```python
print("Magnitudes for a1:")
a1m = (np.real(a1)**2 + np.imag(a1)**2)**(1/2)
print(a1m)
print("\nMagnitudes for a2:")
a2m = (np.real(a2)**2 + np.imag(a2)**2)**(1/2)
print(a2m)
```

```python
print("Highest magnitude among a1m:")
print(math.floor(a1m.max() * 10)/10)
print("\nHighest magnitude among a2m:")
print(math.floor(a2m.max()*10)/10)
```
