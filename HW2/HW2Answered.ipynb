```python
N1 = 9
n1 = np.arange(-4, 5)
k1 = np.arange(-4, 5)
a1k = np.zeros(N1, dtype = complex)
x1 = np.zeros(N1, dtype = complex)
for k in k1:
    a1k[k+4] = 1/9 * np.sum((n1+2*np.exp(3*n1))*np.heaviside(n1-2, 1)*np.exp(-1j*2*np.pi*k*n1/9))
for n in n1:
    x1[n+4] = np.sum(a1k[k1+4]*np.exp(1j*2*np.pi*k1*n/9))


N2 = 11
n2 = np.arange(-5, 6)
k2 = np.arange(-5, 6)
a2k = np.zeros(N2, dtype = complex)
x2 = np.zeros(N2, dtype = complex)
for k in k2:
    a2k[k+5] = 1/11 * np.sum(n2**2 * np.heaviside(-4-n2, 1) * np.exp(-1j*2*np.pi*k*n2/11))
for n in n2:
    x2[n+5] = np.sum(a2k[k2+5]*np.exp(1j*2*np.pi*k2*n/11))


N3 = 7
n3 = np.arange(-3, 4)
k3 = np.arange(-3, 4)
a3k = np.zeros(N3, dtype = complex)
x3 = np.zeros(N3, dtype = complex)
for k in k3:
    a3k[k+3] = 1/7 * np.sum(((n3-2)*np.exp(n3-2) + (n3-1)*np.exp(n3-1) + n3*np.exp(n3) + (n3+1)*np.exp(n3+1) + (n3+2)*np.exp(n3+2))*np.exp(-1j*2*np.pi*n3*k/7))
for n in n3:
    x3[n+3] = np.sum(a3k[k3+3]*np.exp(1j*2*np.pi*k3*n/7))
    

fig, ax = plt.subplots(3, 1, figsize=(10, 12))    
ax[0].stem(n1, np.real(x1))
ax[1].stem(n2, np.real(x2))
ax[2].stem(n3, np.real(x3))
ax[0].set_title("x1[n]")
ax[1].set_title("x2[n]")
ax[2].set_title("x3[n]")
fig.tight_layout()
plt.show()
```
