# Question 1.a

```python
n = np.arange(-100, 100)
t = np.linspace(-100, 100, 10000)

u = np.vectorize(lambda n: int(n >= 0))
imp = np.vectorize(lambda n: int(n == 0))

# x1[n] = 5e^j(7pin/8+pi/3)

x1 = lambda n: 5*np.exp(1j*(7*np.pi*n/8 + np.pi/3))

plt.xlabel("n")
plt.ylabel("real x1[n]")
plt.stem(n, np.real(x1(n)))
plt.title(r'Plot of real part of DT signal $x[n]= 5 e^{j(7\pi n/8 + \pi/3)}$')
plt.show()
plt.xlabel("n")
plt.ylabel("imaginary x1[n]")
plt.stem(n, np.imag(x1(n)))
plt.title(r'Plot of imaginary part of DT signal $x[n]= 5 e^{j(7\pi n/8 + \pi/3)}$')
plt.show()

# x[n] = (0.2n)u[n+2]-(0.2n)u[2-n]+u[5]

x2 = lambda n: 0.2*n*u(n+2) - 0.2*n*u(2-n) + u(5)

plt.xlabel("n")
plt.ylabel("x2[n]")
plt.stem(n, x2(n))
plt.title(r'Plot of DT signal $x[n]= (0.2n)u(n+2)-(0.2n)u(2-n)+u(5)$')
plt.show()

# x[n] = sum[m=0, infinity](delta[n-m])

x3 = lambda n: imp(n<0)

plt.xlabel("n")
plt.ylabel("x3[n]")
plt.stem(n, x3(n))
plt.title(r'Plot of DT signal $x[n]= \sum \delta(n-m)$')
plt.show()

# x(t) = t+1 / 1 /  -t+3 / 0

x4 = lambda t: np.piecewise(t, [(t >= -1) & (t <= 0) , (t >= 0) & (t <= 2), (t >= 2) & (t <= 3), (t <= -1) | (t >= 3)], [lambda t: t+1, 1, lambda t: -t+3, 0])

plt.xlabel("t")
plt.ylabel("x4(t)")
plt.plot(t, x4(t))
plt.show()

# x(t) = 5sin(2pit)cos(pit-8)

x5 = lambda t: 5*np.sin(2*np.pi*t)*np.cos(np.pi*t-8)

plt.xlabel("t")
plt.ylabel("x5(t)")
plt.plot(t, x5(t))
plt.title(r'Plot of CT signal $x(t)= 5\sin(2\pi t)\cos(\pi -8) $')
plt.show()
```

To determine whether the system $ y[n] = a_{0}x[n] + a_{1}x[n-1] + a_{2}x[n-2] + a_{3}x[n-3] $ is BIBO stable we can its impulse response. For a LTI system to be BIBO stable, the absolute sum of its impulse response for n from negative infinity to positive infinity must be finite. If the system is BIBO stable, $$\sum_{n=-\infty}^{\infty}h(n) \not= \infty.$$
To obtain impulse response of the system, we can replace $x[n]$ with $\delta(n)$. Then, the result of the calculation gives us $$h(n) = a_{0}\delta(n) + a_{1}\delta(n-1) + a_{2}\delta(n-2) + a_{3}\delta(n-3).$$ 
Carrying out summation, we acquire result $\sum_{n=-\infty}^{\infty}h(n) = a_{0} + a_{1} + a_{2} + a_{3}$. Given that the coefficients are constant, i.e. not dependent on n, then the system is BIBO stable.
